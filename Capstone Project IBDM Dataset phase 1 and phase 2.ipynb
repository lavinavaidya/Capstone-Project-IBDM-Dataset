# Webscraping of IMBd movies detail 

### Table 1

```python
#install and import required libraries.
!pip install bs4
!pip install requests
from bs4 import BeautifulSoup
import requests
import pandas as pd
import csv
import numpy as np
from time import sleep
from random import randint
```

```python
# information to be in english language
header={"Accept=language":"en-US,en;q=0.5"}
```

```python
# extracting information of pages
pages=np.arange(1,1749,50)
```

```python
# Create empty lists for the data
Sno = []
Movie_Name = []
Director_Name = []
Director1 = []
Director2 = []
Director3 = []
Duration = []
Year = []
Ratings = []
Metascore = []
```

```python
#creating a for loop for required pages
for page in pages:
    #getting required url
    page=requests.get("https://www.imdb.com/search/title/?title_type=feature&num_votes=25000,&genres=action&sort=user_rating,desc&start="+str(page)+"&ref_=adv_nxt")
    soup = BeautifulSoup(page.text, 'html.parser')
    # Find movie data on the current page
    movie_data = soup.find_all('div', attrs={'class': 'lister-item-content'})
    sleep(randint(2,8))
    # Iterate over each movie on the current page
    for movie in movie_data:
        # Extract serial number
        Sno.append(len(Sno) + 1)
        
        # Extract movie name
        Movie_Name.append(movie.h3.a.text)
        
        # Extract director names
        directors = movie.find_all('p')[2].find_all('a')
        director_names = [director.text for director in directors][:-4]
        Director_Name.append(director_names)
         # Assign values to Director1, Director2, and Director3
        director_count = len(director_names)
        Director1.append(director_names[0] if director_count > 0 else "-")
        Director2.append(director_names[1] if director_count > 1 else "-")
        Director3.append(director_names[2] if director_count > 2 else "-")

        # Extract duration
        Duration.append(movie.find('span', attrs={'class': 'runtime'}).text)
        
        # Extract year
        Year.append(movie.find('span', attrs={'class': 'lister-item-year'}).text.strip('()'))
        
        # Extract rating
        Ratings.append(movie.strong.text)
        
        # Extract metascore
        metascore = movie.find('span', attrs={'class': 'metascore'})
        Metascore.append(metascore.text if metascore else "-")
    
```

```python
# Create and print a DataFrame from the scraped data
df = pd.DataFrame({'Sno': Sno,'Movie Name': Movie_Name,'Director1': Director1,'Director2': Director2,'Director3': Director3,'Duration': Duration,'Year': Year,'Ratings': Ratings,'Metascore': Metascore})
```

```python
# showing dataframe
df
```

```python
# saving file to csv
df.to_csv('Table1_IMBd.csv', index=False)
```

### Table 2

```python
# Making blank lists for required data
Movie_Name = []
star_list =[]
star1 = []
star2 = []
star3 = []
star4 = []
votes = []
Genre1 = []
Genre2 = []
Genre3 = []
Gross_collection = []
popularity = []
Certification = []
for page in pages:
    #getting required url
    page=requests.get("https://www.imdb.com/search/title/?title_type=feature&num_votes=25000,&genres=action&sort=user_rating,desc&start="+str(page)+"&ref_=adv_nxt")
    soup = BeautifulSoup(page.text, 'html.parser')
    # Find movie data on the current page
    movie_data = soup.find_all('div', attrs={'class': 'lister-item-content'})
    sleep(randint(2,8))
    # Iterate over each movie on the current page
    for movie in movie_data:
        # Extract movie name
        movie_name = movie.find('h3').find('a').text.strip()
        Movie_Name.append(movie_name)
        
        # Extract stars of the movie
        star = movie.find_all('p')[2].find_all('a')
        star_lis = [star.text for star in star]
        if len(star_lis) >= 4:
            star1.append(star_lis[-4])
            star2.append(star_lis[-3])
            star3.append(star_lis[-2])
            star4.append(star_lis[-1])
        elif len(star_lis) == 3:
            star1.append(star_lis[-3])
            star2.append(star_lis[-2])
            star3.append(star_lis[-1])
            star4.append("-")
        elif len(star_lis) == 2:
            star1.append(star_lis[-2])
            star2.append(star_lis[-1])
            star3.append("-")
            star4.append("-")
        elif len(star_lis) == 1:
            star1.append(star_lis[-1])
            star2.append("-")
            star3.append("-")
            star4.append("-")
        else:
            star1.append("-")
            star2.append("-")
            star3.append("-")
            star4.append("-")
        
        # Extract votes and gross collection
        values=soup.find_all('span', attrs={'name': 'nv'})
        vote= values[0].text
        votes.append(vote)
        gross=values[1].text if len(values) >1 else "-"
        Gross_collection.append(gross)
        
        # Extract genre
        genre_elements = movie.find_all('span', class_='genre')
        genres = []
        for genre_element in genre_elements:
            genre_data = genre_element.text.strip()
            genres.extend(genre_data.split(', '))
        # Bifurcate genres into three subfields
        genre1 = None
        genre2 = None
        genre3 = None
        if len(genres) > 0:
            genre1 = genres[0].strip()
        if len(genres) > 1:
            genre2 = genres[1].strip()
        if len(genres) > 2:
            genre3 = genres[2].strip()
        Genre1.append(genre1)
        Genre2.append(genre2)
        Genre3.append(genre3)

        # Extract popularity
        popularity_data = movie.find('span', attrs={'name': 'ir'})
        if popularity_data is not None:
            popularity_text = popularity_data.text.strip()
            popularity.append(popularity_text)
        else:
            popularity.append(None)
        
        # Extract certification
        certification_element = movie.find('span', class_='certificate')
        if certification_element is not None:
            certification_data = certification_element.text.strip()
            Certification.append(certification_data)
        else:
            Certification.append(None)
```

```python
# Create a DataFrame to store the scraped data
df1 = pd.DataFrame({'Movie_Name': Movie_Name,'Star1': star1,'Star2': star2,'Star3': star3,'Star4': star4,'Votes': votes,'Genre1': Genre1,'Genre2': Genre2,'Genre3': Genre3,'Gross_Collection': Gross_collection,'Popularity': popularity,'Certification': Certification})
```

```python
# saving file to csv
df1.to_csv('Table2_IMBd.csv', index=False)
```

### subquery

1) Display all details from both tables where movie names are the same.

```python
# Execute the SQL query 
result = cur.execute("SELECT *FROM IMDB_1 JOIN IMDB_2 ON IMDB_1.[Movie Name] = IMDB_2.Movie_Name")
# Fetch all rows
row = result.fetchall()
# Create a DataFrame from the rows and column names
df = pd.DataFrame(row, columns=[column[0] for column in result.description])
df
```

2)   Display all movie names, Director, ratings, and gross collection where the genre is action.

```python
# Execute the SQL query 
result = cur.execute("SELECT IMDB_1.[Movie Name], IMDB_1.Director1, IMDB_1.Director2, IMDB_1.Director3, IMDB_1.Ratings, IMDB_2.Gross_Collection FROM IMDB_1 JOIN IMDB_2 ON IMDB_1.[Movie Name] = IMDB_2.Movie_Name WHERE IMDB_2.Genre1 = 'Action' OR IMDB_2.Genre2 = 'Action' OR IMDB_2.Genre3 = 'Action' ")
# Fetch all rows
rows = result.fetchall()
# Create a DataFrame from the rows and column names
df = pd.DataFrame(rows, columns=[column[0] for column in result.description])
df
```
