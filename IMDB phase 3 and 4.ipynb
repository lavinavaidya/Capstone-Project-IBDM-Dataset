Here we find change in datatype in column metascore, gross collection and votes that is due to present of NaN values

```python
# Replace 'Nan' with NaN in 'Gross Collection' column
df['Gross Collection'] = df['Gross Collection'].replace('Nan', np.nan)

# Replace 'Nan' with NaN in 'votes' column
df['votes'] = df['votes'].replace('Nan', np.nan)

# Replace 'Nan' with NaN in 'Metascore' column
df['Metascore'] = df['Metascore'].replace('Nan', np.nan)
```

```python
# Remove commas from the 'votes' column
df['votes'] = df['votes'].str.replace(',', '')


# Convert the 'votes' column  to int
df['votes'] = df['votes'].astype(int)
```

```python
# Replace any non-numeric values or NaN with a default value or appropriate value based on your dataset
df['Gross Collection'] = pd.to_numeric(df['Gross Collection'], errors='coerce').fillna(0)

# Convert the column to float type
df['Gross Collection'] = df['Gross Collection'].astype(float)
```

```python
# checking information of dataset
df.info()
```

```python
# Fill missing values with the respective column's median
df['Metascore'] = df['Metascore'].fillna(df['Metascore'].median)

# Fill missing values with the respective column's median
df['Gross Collection'] = df['Gross Collection'].fillna(df['Gross Collection'].median)
```

```python
# Replace any non-finite values (NaN or inf) with a default value or appropriate value based on your dataset
df['Metascore'] = df['Metascore'].replace([np.inf, -np.inf], np.nan).fillna(0)

# Convert the column to numeric type
df['Metascore'] = pd.to_numeric(df['Metascore'], errors='coerce')

# Convert the column to integer type
df['Metascore'] = df['Metascore'].astype(int)
```

### Model Building

```python
# importing necessary libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
lr=LinearRegression()
```

```python
#finding best random state
maxAccu=0
maxRS=0
for i in range(1,200):
    x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=.2, random_state =i)
    lr.fit(x_train,y_train)
    pred_train=lr.predict(x_train)
    pred_test=lr.predict(x_test)
    acc=r2_score(y_test,pred_test)
    if acc>maxAccu:
        maxAccu=acc
        maxRS=i
print("Best accuracy is ",maxAccu," on Random_state ",maxRS)
```
